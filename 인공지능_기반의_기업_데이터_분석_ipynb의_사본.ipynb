```python
# 독립변수feature와 종속변수label 선택하기
x = data_icecream[['temperature']]
y = data_icecream[['icecream_sales']]

# 분석기법 선택하여 학습하기
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x, y)

# 학습 성능 (예측력) 확인하기 : 예측 정확도 R²
print('예측설명력:', format(model.score(x,y)) )

# from sklearn.metrics import mean_squared_error
# print('예측오차:', format(mean_squared_error(y, model.predict(x), squared=False))) 
```

# Deep Learning - 범주형
 - 고객이탈 관련 데이터

```python
import tensorflow as tf
import numpy as np    

tf.random.set_seed(0)             # seed값 설정: 매번 실행시 같은 결과값이 나오게 함. 단 라이브러리 내부 랜덤 함수가 추가로 작동할 경우 결과값이 다를 수 있음.
np.random.seed(0)                 

from keras.models import Sequential     
from keras.layers import Dense   

# 통신사 고객이탈 데이터: data_churn 
x=data_churn.drop(columns='churn')
y=data_churn['churn']

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=1) 

model = Sequential()   
model.add(Dense(32,  activation='relu'))             # Dense (Fully Connected Layer) 레이어의 입력과 출력을 빠짐없이 모두 연결함. # Node (neuron) 수   
model.add(Dense(1, activation='sigmoid'))                                # sigmoid: 0과 1의 종속변수(label)에서는 마지막 연산 결과를 0~1의 확률값으로 나타내기 위해 activation 함수로서 시그모이드(sigmoid)함수를 사용함, 연속형 종속변수에서는 activation='relu'.   
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])       
model.fit(train_x, train_y, epochs=10, validation_split=0.25, verbose=1) # epochs:  만약 데이터 샘플이 100개 있을 때, 이를 모두 1회 학습하면 1 epochs    # batch_size를 정해해주시 않으면 기본값 32로 진행함

model.evaluate(test_x, test_y)   
  # 딥러닝에서는 accuracy를 계산하는데 model.evaluate를 model.score 대신 사용함
  # 첫번째 결과값은 loss, 두번째 결과값은 accuracy를 의미함
  # loss, acc = model.evaluate(test_x, test_y, verbose=0)   

pred_y=(model.predict(test_x)  >= 0.5).astype(int)           # True는 1로, False는 0으로 출력하기 위한 명령어

accuracy=accuracy_score(test_y, pred_y)
precision=precision_score(test_y, pred_y)   
recall= recall_score(test_y, pred_y)
f1= f1_score(test_y, pred_y) 

accuracy, precision, recall, f1
```

```python
model_name='DeepLearning 1'
output2=output2.append({ 'Name':model_name, 'Accuracy':accuracy, 'Precison':precision, 'Recall':recall, 'F1':f1}, ignore_index=True )
display(output2)
```
